```python
import pandas as pd
import numpy as np
import re

df.columns = df.columns.str.strip().str.lower()
df = df.loc[:, ~df.columns.duplicated()]
df.head()
```

```python
def clean_base_name(col):
    col = re.sub(r'_i_.*', '', col)
    col = re.sub(r'_r_.*', '', col)
    col = re.sub(r'_col\._\(\d+\)', '', col)
    return col

base_names = {col: clean_base_name(col) for col in df.columns}
```

```python
from collections import defaultdict

groups = defaultdict(list)

for col, base in base_names.items():
    groups[base].append(col)

groups = {k: v for k, v in groups.items() if len(v) > 1}
groups
```

```python
for base, cols in groups.items():
    
    incidence_cols = [c for c in cols if '_i_' in c]
    rate_cols = [c for c in cols if '_r_' in c]
    
    if len(incidence_cols) > 1:
        df[base + '_total_incidence'] = df[incidence_cols].sum(axis=1)
        df = df.drop(columns=incidence_cols)
    
    if len(rate_cols) > 1:
        df[base + '_avg_rate'] = df[rate_cols].mean(axis=1)
        df = df.drop(columns=rate_cols)
    
    if len(incidence_cols) == 1:
        df = df.rename(columns={incidence_cols[0]: base + '_incidence'})
    
    if len(rate_cols) == 1:
        df = df.rename(columns={rate_cols[0]: base + '_rate'})
```

```python
if 'population' in df.columns:
    for col in df.columns:
        if col.endswith('_incidence'):
            df[col.replace('_incidence', '_rate_recomputed')] = (df[col] / df['population']) * 100000
```

```python
incidence_cols = [col for col in df.columns if col.endswith('_incidence')]
```

```python
macro_map = {
    "crime_against_women": [
        "crime_against_women",
        "dowry",
        "domestic",
        "rape",
        "sexual",
        "cruelty_by_husband"
    ],
    
    "crime_against_children": [
        "crime_against_children",
        "pocso",
        "child_marriage",
        "child_labour"
    ],
    
    "economic_offences": [
        "cheating",
        "forgery",
        "fraud",
        "breach_of_trust",
        "economic"
    ],
    
    "violent_crime": [
        "murder",
        "culpable_homicide",
        "attempt_to_murder",
        "kidnapping",
        "abduction",
        "rioting",
        "assault"
    ]
}
for macro, keywords in macro_map.items():
    
    relevant_cols = []
    
    for col in incidence_cols:
        for key in keywords:
            if key in col:
                relevant_cols.append(col)
                break
    
    if len(relevant_cols) > 0:
        df[macro + "_total"] = df[relevant_cols].sum(axis=1)
```

```python
target_column = "overall_crime_index"
numeric_df = df.select_dtypes(include="number")

X = numeric_df.drop(columns=[target_column])
y = numeric_df[target_column]
corr_values = X.corrwith(y).abs()

selected_features_corr = corr_values[corr_values >= 0.4].index.tolist()

X_corr_selected = X[selected_features_corr]

print("Selected by Correlation:")
print(selected_features_corr)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=42)
model.fit(X_corr_selected, y)

importances = pd.Series(model.feature_importances_, 
                        index=X_corr_selected.columns)

important_features_model = importances.sort_values(ascending=False)

important_features_model
```

```python
top_features = important_features_model.head(10).index.tolist()

X_model_selected = X_corr_selected[top_features]

print("Top Model-Based Features:")
print(top_features)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_model_selected)
```

```python
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

print("Original Features:", X_model_selected.shape[1])
print("Reduced Components:", X_pca.shape[1])
```

```python
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
```

```python
df_pca = pd.DataFrame(
    X_pca,
    columns=["crime_dimension_1", "crime_dimension_2"]
)

df_final = pd.concat([df.reset_index(drop=True), df_pca], axis=1)
```
